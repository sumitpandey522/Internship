```python
# import libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import time


# connect to the driver

driver = webdriver.Chrome(r'D:/SUMIT/chromedriver_win32/chromedriver.exe')

# opening url

url = "https://www.azquotes.com/"
driver.get(url)

driver.maximize_window()
time.sleep(3)
```

```python
# clicking on top quotes

top= driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top.click()
```

```python
# scraping a) Quote b) Author c) Type Of Quote

Quotes_list=[]
Authors_list=[]
Quote_type_list=[]

start=0
end=2   # for scrape all 1000 quotes replace the value 2 with 10

for i in range(start,end):
    quote= driver.find_elements(By.XPATH,"//a[@class='title']")
    
    author= driver.find_elements(By.CLASS_NAME,"author")
    
    types= driver.find_elements(By.CLASS_NAME,"tags")
    
    
    
    for qt in quote:
        Quotes_list.append(qt.text)
        
    for au in author:
        Authors_list.append(au.text)
        
    for tl in types:
        Quote_type_list.append(tl.text)
        
    
        
        
    nxt_button= driver.find_element(By.XPATH,"//li[@class='next']")
    nxt_button.click()
    time.sleep(5)

driver.close()
```

```python
print(len(Quotes_list),
len(Authors_list),
len(Quote_type_list))

df=pd.DataFrame()

df["QUOTES"]=Quotes_list

df["AUTHOR"]=Authors_list

df["TYPE_OF_QUOTES"]=Quote_type_list

df
```
